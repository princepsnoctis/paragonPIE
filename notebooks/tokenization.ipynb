Collect all unique characters in the dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv", dtype=str)

df_as_text = df.astype(str).agg("".join, axis=1).str.cat()

vocab = sorted(set(df_as_text))

vocab
```

Add special control tokens

```python
PAD = "<PAD>"
BOS = "<BOS>"
EOS = "<EOS>"
UNK = "<UNK>"

vocab = [PAD, BOS, EOS, UNK] + vocab

vocab
```

Assign each character an integer ID

```python
c2i = {c:i for i,c in enumerate(vocab)}
i2c = {i:c for i,c in enumerate(vocab)}

PAD_i = c2i[PAD]
BOS_i = c2i[BOS]
EOS_i = c2i[EOS]
UNK_i = c2i[UNK]

i2c[c2i["Z"]]
```
